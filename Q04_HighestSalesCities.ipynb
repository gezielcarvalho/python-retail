# Exploratory Data Analysis (EDA). Retail Industry
## Top 10 Cities with Highest Total Sales

```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

### Loading the dataset

```python
# Loads the dataset
dataframe_retail = pd.read_csv('data/dataset.csv')
```

### Exploratory Analysis

```python
# Checking for duplicated data
dataframe_retail[dataframe_retail.duplicated()]
```

```python
# Checking for missing values
dataframe_retail.isnull().sum()
```

### Identify the top 10 cities with the highest total sales and visualize the results using a bar chart


```python
# Get the top 10 cities
dataframe_retail_total_per_state = dataframe_retail.groupby('City')['TotalOrderValue'].sum().reset_index().sort_values(by = 'TotalOrderValue', ascending = False).head(10)
```

```python
dataframe_retail_total_per_state
```

```python
# Create plot.
plt.figure(figsize = (16,6))
sns.set_palette('coolwarm')
sns.barplot(
    data = dataframe_retail_total_per_state, 
    x = 'TotalOrderValue', 
    y = 'City').set(title = "Top 10 Cities with Highest Total Sales")
plt.xticks(rotation = 60)
plt.show()
```
